```python
test = og_df
test.replace({'total_cloud_cover_sfc': 100}, 11, inplace=True)
sn.histplot(test['total_cloud_cover_sfc'])
```

```python
label = test['generated_power_kw']
#df = og_df
test.drop('generated_power_kw', axis=1, inplace=True)
test
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
test = np.array(scaler.fit_transform(test))
#test = pd.DataFrame(test)
test
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(test, label, test_size=0.2)
```

```python
from sklearn.preprocessing import PolynomialFeatures

Poly = PolynomialFeatures(degree=2)
x_train = Poly.fit_transform(x_train)
x_test = Poly.fit_transform(x_test)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
model = Sequential()
model.add(layers.Dense(32, activation="relu", input_shape=(x_train.shape[1],)))
model.add(layers.Dense(16, activation="relu"))
model.add(layers.Dense(1))
print(model.summary())
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'mse',
    metrics = ['accuracy']
)
```

```python
history = model.fit(x_train,
                    y_train,
                    validation_split = 0.2,
                    batch_size = 50,
                    epochs = 150,
                    callbacks = [tensorflow.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10,
                                                                  restore_best_weights = True)]
                    )
```

```python
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss')
```

```python
from sklearn.linear_model import LinearRegression
poly_reg_model = LinearRegression()
poly_reg_model.fit(x_train, y_train)
predict = poly_reg_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, predict)
```

```python
y_pred = model.predict(x_test)  # Get model predictions on the test set
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```
